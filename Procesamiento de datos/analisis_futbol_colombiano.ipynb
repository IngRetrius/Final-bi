**Importación de librerías y carga de datos**

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime

# Configuración de visualización
plt.style.use('ggplot')
sns.set(style="whitegrid")
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# Cargar los datos unificados
# Ajusta la ruta según donde tengas tu archivo
ruta_datos = "data/jugadores_unificados_cinco_torneos.csv"
df = pd.read_csv(ruta_datos)

# Mostrar información general del dataset
print(f"Dimensiones del dataset: {df.shape}")
print(f"\nPrimeras 5 filas:")
display(df.head())

# Información sobre tipos de datos y valores no nulos
print("\nInformación del dataset:")
display(df.info())

# Estadísticas descriptivas básicas
print("\nEstadísticas descriptivas:")
pd.set_option('display.max_columns', None)  # Muestra todas las columnas
pd.set_option('display.width', 1000)        # Ancho amplio para evitar cortes
display(df.describe().transpose()) 
```

**LIMPIEZA Y EXPLORACIÓN DE VALORES FALTANTES**

```python
# Verificar valores nulos por columna
print("Valores nulos por columna:")
display(df.isnull().sum())

# Porcentaje de valores nulos
print("\nPorcentaje de valores nulos por columna:")
display((df.isnull().sum() / len(df) * 100).sort_values(ascending=False))

# Verificar tipos de datos
print("\nTipos de datos:")
display(df.dtypes)

# Convertir columnas numéricas si es necesario
columnas_numericas = ['Goals', 'Assists', 'Succ. dribbles', 'Tackles', 
                     'Accurate passes %', 'Total shots', 'Goal conversion %']

for col in columnas_numericas:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Análisis de duplicados
print(f"\nRegistros duplicados: {df.duplicated().sum()}")

# Distribución de jugadores por torneo
print("\nDistribución de registros por torneo:")
display(df['Torneo'].value_counts())

# Distribución de jugadores por equipo
print("\nTop 10 equipos por número de jugadores:")
display(df['Team'].value_counts().head(10))
```

**FILTRADO DE JUGADORES PARA ANÁLISIS DE EVOLUCIÓN**

```python
# Obtener lista de jugadores que participaron en el torneo más reciente (Apertura 2025A)
jugadores_2025A = df[df['Torneo'] == 'Apertura 2025A']['Name'].unique()
print(f"Número de jugadores en Apertura 2025A: {len(jugadores_2025A)}")

# Filtrar el dataset para mantener solo estos jugadores
df_evolucion = df[df['Name'].isin(jugadores_2025A)]

# Verificar número de registros por torneo para estos jugadores
print("\nDistribución por torneo de los jugadores seleccionados:")
display(df_evolucion['Torneo'].value_counts())

# Verificar cuántos jugadores tienen datos en los diferentes torneos
jugadores_por_torneo = df_evolucion.groupby('Name')['Torneo'].nunique()
print(f"\nJugadores con datos en los cinco torneos: {(jugadores_por_torneo == 5).sum()}")
print(f"Jugadores con datos en cuatro torneos: {(jugadores_por_torneo == 4).sum()}")
print(f"Jugadores con datos en tres torneos: {(jugadores_por_torneo == 3).sum()}")
print(f"Jugadores con datos en dos torneos: {(jugadores_por_torneo == 2).sum()}")
print(f"Jugadores con datos en un solo torneo: {(jugadores_por_torneo == 1).sum()}")

# Crear un nuevo dataframe con la evolución completa
# Ordenado primero por nombre y luego por torneo según el orden especificado
orden_torneos = {
    'Apertura 2025A': 0,   # Más reciente
    'Clausura 2024B': 1, 
    'Apertura 2024A': 2,
    'Clausura 2023B': 3,
    'Apertura 2023A': 4    # Más antiguo
}
df_evolucion['orden_torneo'] = df_evolucion['Torneo'].map(orden_torneos)
df_evolucion = df_evolucion.sort_values(['Name', 'orden_torneo'])
df_evolucion = df_evolucion.drop('orden_torneo', axis=1)

# Mostrar las primeras filas del dataset filtrado
print("\nPrimeras filas del dataset filtrado y ordenado:")
display(df_evolucion.head(15))  # Muestra 15 filas para ver múltiples jugadores

# Guardar el dataset filtrado 
df_evolucion.to_csv("data/evolucion_jugadores_2025A.csv", index=False)

# Opcional: Generar estadísticas sobre la continuidad de los jugadores
continuidad_jugadores = pd.DataFrame({
    'num_torneos': jugadores_por_torneo
})
print("\nEstadísticas de continuidad de jugadores:")
display(continuidad_jugadores['num_torneos'].value_counts().sort_index())
```

**Graficar evolución de goles para los jugadores más destacados**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset
df_evolucion = pd.read_csv("data/evolucion_jugadores_2025A.csv")

# Asegurar orden cronológico
orden_cronologico = ['Apertura 2023A', 'Clausura 2023B', 'Apertura 2024A', 'Clausura 2024B', 'Apertura 2025A']
df_evolucion['Torneo'] = pd.Categorical(df_evolucion['Torneo'], categories=orden_cronologico, ordered=True)

# Encontrar el goleador de cada torneo
goleadores_por_torneo = df_evolucion.loc[df_evolucion.groupby('Torneo')['Goals'].idxmax()]

# Lista de jugadores destacados (puede haber repetidos)
jugadores_destacados = goleadores_por_torneo['Name'].unique().tolist()
print(f"Jugadores destacados por torneo: {jugadores_destacados}")

# Filtrar datos de esos jugadores
df_top = df_evolucion[df_evolucion['Name'].isin(jugadores_destacados)]

# Crear gráfico de evolución
plt.figure(figsize=(14, 8))

for jugador in jugadores_destacados:
    datos_jugador = df_top[df_top['Name'] == jugador].sort_values('Torneo')
    plt.plot(datos_jugador['Torneo'], datos_jugador['Goals'], marker='o', linewidth=2, label=jugador)

plt.title('Evolución de goles de los máximos goleadores por torneo', fontsize=16)
plt.ylabel('Goles por torneo', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

**Análisis de Evolución de Defensas en el Fútbol**

```python
# Calcular jugadores por torneo
jugadores_por_torneo = df_evolucion.groupby('Name')['Torneo'].nunique()

# Seleccionar jugadores con datos en los 5 torneos
jugadores_completos = jugadores_por_torneo[jugadores_por_torneo == 5].index.tolist()
print(f"Número de jugadores con datos en los 5 torneos: {len(jugadores_completos)}")

# Primero, verificar y corregir posibles duplicados en mismo torneo
df_evolucion_corregido = df_evolucion.copy()
duplicados_count = 0

# Para cada jugador, verificar si tiene duplicados en algún torneo
for jugador in jugadores_completos:
    for torneo in df_evolucion['Torneo'].unique():
        # Contar registros para este jugador y torneo
        registros = df_evolucion[(df_evolucion['Name'] == jugador) & 
                              (df_evolucion['Torneo'] == torneo)]
        
        if len(registros) > 1:
            duplicados_count += 1
            # Eliminar duplicados manteniendo el primero (o puedes usar otras estrategias)
            indices_a_eliminar = registros.index[1:]
            df_evolucion_corregido = df_evolucion_corregido.drop(indices_a_eliminar)

print(f"Se encontraron y corrigieron {duplicados_count} casos de jugadores con datos duplicados en un mismo torneo")

# Crear un puntaje defensivo combinado (suma de tackles, intercepciones y despejes)
df_defensa = df_evolucion_corregido[df_evolucion_corregido['Name'].isin(jugadores_completos)].copy()
df_defensa['Defensive_Score'] = df_defensa['Tackles'].fillna(0) + df_defensa['Interceptions'].fillna(0) + df_defensa['Clearances'].fillna(0)

# Calcular el puntaje defensivo total por jugador
top_defensas = df_defensa.groupby('Name')['Defensive_Score'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 defensas basado en métricas defensivas combinadas:")
print(top_defensas)

# Establecer orden cronológico correcto
orden_cronologico = ['Apertura 2023A', 'Clausura 2023B', 'Apertura 2024A', 'Clausura 2024B', 'Apertura 2025A']

# Función para graficar evolución de una métrica para los mejores jugadores
def graficar_evolucion(metrica, titulo, ylabel):
    plt.figure(figsize=(14, 8))
    
    for jugador in top_defensas.index:
        # Filtrar datos del jugador (usando datos corregidos)
        datos_jugador = df_evolucion_corregido[df_evolucion_corregido['Name'] == jugador]
        
        # Verificar que solo hay un registro por torneo
        for torneo in orden_cronologico:
            count = len(datos_jugador[datos_jugador['Torneo'] == torneo])
            if count > 1:
                print(f"⚠️ {jugador} todavía tiene {count} registros en {torneo}")
        
        # Crear un DataFrame ordenado para este jugador
        datos_ordenados = []
        for torneo in orden_cronologico:
            fila = datos_jugador[datos_jugador['Torneo'] == torneo]
            if not fila.empty:
                datos_ordenados.append(fila)
        
        if datos_ordenados:
            datos_jugador_ordenado = pd.concat(datos_ordenados)
            plt.plot(datos_jugador_ordenado['Torneo'], datos_jugador_ordenado[metrica], 'o-', 
                    label=f"{jugador} ({datos_jugador_ordenado['Team'].iloc[0]})", 
                    linewidth=2, markersize=8)
    
    plt.title(titulo, fontsize=16)
    plt.ylabel(ylabel, fontsize=14)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Gráficos de evolución para diferentes métricas
graficar_evolucion('Tackles', 'Evolución de tackles de los 5 mejores defensas a lo largo de los torneos', 'Tackles por torneo')
graficar_evolucion('Interceptions', 'Evolución de intercepciones de los 5 mejores defensas a lo largo de los torneos', 'Intercepciones por torneo')
graficar_evolucion('Clearances', 'Evolución de despejes de los 5 mejores defensas a lo largo de los torneos', 'Despejes por torneo')


```

**Analisis de arqueros**

```python
# Configuración de visualización
plt.style.use('ggplot')
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 12

# Identificar verdaderos arqueros basado en estadísticas relevantes
verdaderos_arqueros_mask = (
    df_evolucion['Saves'].notna() & 
    (df_evolucion['Goals'].fillna(0) < 2) & 
    (df_evolucion['Succ. dribbles'].fillna(0) < 4)
)

df_arqueros = df_evolucion[verdaderos_arqueros_mask].copy()
print(f"Se identificaron {df_arqueros['Name'].nunique()} arqueros en el dataset")

# Verificar si hay arqueros con el mismo nombre en diferentes equipos
for arquero in df_arqueros['Name'].unique():
    equipos = df_arqueros[df_arqueros['Name'] == arquero]['Team'].unique()
    if len(equipos) > 1:
        print(f"Arquero {arquero} aparece en múltiples equipos: {equipos}")
        # Renombrar para distinguir por equipo
        for equipo in equipos:
            mask = (df_arqueros['Name'] == arquero) & (df_arqueros['Team'] == equipo)
            df_arqueros.loc[mask, 'Name'] = f"{arquero} ({equipo})"

# Calcular torneos por arquero
arqueros_por_torneo = df_arqueros.groupby('Name')['Torneo'].nunique()

# Filtrar arqueros con datos en al menos 3 torneos
arqueros_consistentes = arqueros_por_torneo[arqueros_por_torneo >= 3].index.tolist()
print(f"Arqueros con datos en al menos 3 torneos: {len(arqueros_consistentes)}")

# Crear un score combinado para arqueros
df_arqueros_consistentes = df_arqueros[df_arqueros['Name'].isin(arqueros_consistentes)].copy()
df_arqueros_consistentes['GK_Score'] = (
    df_arqueros_consistentes['Saves'].fillna(0) +                 # Atajadas
    df_arqueros_consistentes['Clean sheet'].fillna(0) * 10 +      # Vallas invictas (ponderadas x10)
    df_arqueros_consistentes['Penalties saved'].fillna(0) * 5 +   # Penales atajados (ponderados x5)
    df_arqueros_consistentes['Saves from inside box'].fillna(0)   # Atajadas dentro del área
)

# Calcular puntaje total y seleccionar top 5
top_arqueros = df_arqueros_consistentes.groupby('Name')['GK_Score'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 arqueros basado en métricas combinadas:")
print(top_arqueros)

# Establecer orden cronológico
orden_cronologico = ['Apertura 2023A', 'Clausura 2023B', 'Apertura 2024A', 'Clausura 2024B', 'Apertura 2025A']

# Graficar evolución de atajadas para los mejores arqueros
plt.figure(figsize=(14, 8))

# Colores distintivos para cada arquero
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

for i, arquero in enumerate(top_arqueros.index):
    # Filtrar datos del arquero
    datos_arquero = df_arqueros_consistentes[df_arqueros_consistentes['Name'] == arquero]
    
    # Crear un DataFrame ordenado
    datos_ordenados = []
    for torneo in orden_cronologico:
        fila = datos_arquero[datos_arquero['Torneo'] == torneo]
        if not fila.empty:
            datos_ordenados.append(fila)
    
    if datos_ordenados:
        datos_arquero_ordenado = pd.concat(datos_ordenados)
        equipo = datos_arquero_ordenado['Team'].iloc[0]
        
        # Incluir solo el nombre del equipo en la leyenda si no está ya en el nombre
        if '(' in arquero:
            etiqueta = arquero
        else:
            etiqueta = f"{arquero} ({equipo})"
        
        plt.plot(datos_arquero_ordenado['Torneo'], datos_arquero_ordenado['Saves'], 'o-', 
                label=etiqueta, color=colores[i], linewidth=2.5, markersize=9)

plt.title('Evolución de atajadas de los 5 mejores arqueros de la Liga Colombiana', fontsize=16)
plt.ylabel('Atajadas por torneo', fontsize=14)
plt.xlabel('Torneo', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
